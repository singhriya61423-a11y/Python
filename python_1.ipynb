```python
# Step 1: Store information of 5 students
students = [
    {
        "name": "Rohit",
        "age": 21,
        "courses": ("HTML", "SQL"),
        "skills": {"HTML", "SQL"}
    },
    {
        "name": "Aarav",
        "age": 20,
        "courses": ("Python", "HTML"),
        "skills": {"Python", "HTML"}
    },
    {
        "name": "Meha",
        "age": 21,
        "courses": ("Python", "CSS"),
        "skills": {"Python", "CSS"}
    },
    {
        "name": "Kunal",
        "age": 19,
        "courses": ("CSS", "SQL"),
        "skills": {"CSS", "SQL"}
    },
    {
        "name": "Sneha",
        "age": 20,
        "courses": ("SQL", "CSS"),
        "skills": {"SQL", "CSS"}
    }
]

# Step 2: Add one new student
new_student = {
    "name": "Neha",
    "age": 21,
    "courses": ("SQL", "HTML"),
    "skills": {"SQL", "HTML"}
}

students.append(new_student)

# Step 3: Update age of one student
students[0]["age"] = 22   # Rohit's age updated

# Step 4: Display total students
print("Total Students:", len(students))

# Step 5: Students enrolled in Python
print("\nStudents Enrolled in Python:")
for student in students:
    if "Python" in student["courses"]:
        print("-", student["name"])

# Step 6: Unique skills across all students
all_skills = set()
for student in students:
    all_skills.update(student["skills"])

print("\nUnique Skills Across All Students:")
print(all_skills)

# Step 7: Find the oldest student
oldest = students[0]
for student in students:
    if student["age"] > oldest["age"]:
        oldest = student

print("\nOldest Student:")
print("Name:", oldest["name"])
print("Age:", oldest["age"])
```

```python
def even_list(lst):
    result = []
    for n in lst:
        if n % 2 == 0:
            result.append(n)
    return result

print(even_list([1,2,3,4,5,6]))
```
